```python
import time

import pandas as pd
from nilmtk import DataSet
from sklearn.externals import joblib
import matplotlib.pyplot as plt
import numpy as np

SAVED_MODEL = "../../clustering_model/clf.pkl"
UK_DALE = '../../../Datasets/UKDALE/ukdale.h5'
PATH_SIGNAL2VEC = '../../embeddings/signal2vec.csv'
dataset = DataSet(UK_DALE)
clf = joblib.load(SAVED_MODEL)

def get_main_and_submeter_data(start, end, sample_period, buidling):
    global elec, mains, main_power
    dataset.set_window(start=start, end=end)
    elec = dataset.buildings[buidling].elec
    mains = elec.mains()
    main_power = mains.power_series_all_data(sample_period=sample_period)
    df_main = pd.DataFrame(main_power)
    df_main.replace('NaN', 0, inplace=True)
    print('Dataframe of aggregated power consumption')
    print(df_main.head())
    df_appliance_meters = elec.submeters().dataframe_of_meters()
    df_appliance_meters.fillna(0, inplace=True)
    print('\nLength of samples per appliance {}'.format(len(df_appliance_meters)))
    print('Length of samples of aggregated power {}'.format(len(df_main)))
    return df_main, df_appliance_meters
    
def vectorize_sequence(df, filename):
    start = time.time()
    pred = clf.predict(df)
    print('Length of data {}'.format(len(df)))
    print('Length of predicted sequence {}'.format(len(pred)))
    
    plt.plot(pred)
    plt.savefig(filename + '.png')
    plt.show()
    
    vectors = pd.read_csv(PATH_SIGNAL2VEC)
    print('Shape of vectors {}'.format(vectors.shape))
    sequence_of_vectors = list()
    for i in pred:
        sequence_of_vectors.append(vectors[str(i)])

    df_vectorized = pd.DataFrame(sequence_of_vectors)
    df_vectorized = df_vectorized.reset_index(drop=True)
    df_vectorized.to_csv(filename + '.csv', index=False)
    print('Shape of vectorized sequence {}'.format(df_vectorized.shape))
    print("Duration: {}".format(time.time() - start))
```

```python
import os

year = '2014'
month = '2'
end = "{}-28-{}".format(month, year)
start = "{}-1-{}".format(month, year)
sample_period = 6
building = 1

FOLDER_PATH = '/media/christoforos/DATA/signal2vec/vectorized_sequences/{}/{}'.format(year, month)
if not os.path.exists(FOLDER_PATH):
    os.makedirs(FOLDER_PATH)
        
df_main, df_appliances = get_main_and_submeter_data(start, end, sample_period, building)


plt.plot(df_main)
plt.savefig(FOLDER_PATH + '/mains_orig' + '.png')
plt.show()
vectorize_sequence(df_main, FOLDER_PATH + '/mains')

cols = df_appliances.columns
for col in cols:
    plt.plot(df_appliances[col])
    plt.savefig(FOLDER_PATH + '/{}'.format(appliance_name) + '_orig' + '.png')
    plt.show()
    data = np.reshape(df_appliances[col].values, (len(df_appliances), 1))
    appliance_name = elec[col].label() + str(col.instance)
    vectorize_sequence(data, FOLDER_PATH + '/{}'.format(appliance_name))

    
```
