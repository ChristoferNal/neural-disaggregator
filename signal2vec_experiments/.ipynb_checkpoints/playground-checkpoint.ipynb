```python
import pandas as pd
from nilmtk import DataSet
from sklearn.externals import joblib
import matplotlib.pyplot as plt
import numpy as np

SAVED_MODEL = "../clustering_model/clf.pkl"
UK_DALE = '../../Datasets/UKDALE/ukdale.h5'
# END = "20-3-2014"
END = "2-28-2014"
START = "2-1-2014"
DEVICE = "fridge"
sample_period = 6
train_building = 1
clf = joblib.load(SAVED_MODEL)

   
dataset = DataSet(UK_DALE)
dataset.set_window(start=START, end=END)
elec = dataset.buildings[train_building].elec
# meter = elec.submeters()[DEVICE]
mains = elec.mains()
main_power = mains.power_series(sample_period)
vectors = pd.read_csv('../embeddings/signal2vec.csv')
chunk = next(main_power)
df = pd.DataFrame(chunk)
print(df.head())
pred = clf.predict(df)
df = None
print('Length of predected sequence {}'.format(len(pred)))
# plt.plot(pred)
# plt.savefig('clf_10_days.png')
vectors = pd.read_csv('../embeddings/signal2vec.csv')
print(vectors.head())
print(vectors.shape)
sequence_of_vectors = list()
for i in pred:
    sequence_of_vectors.append(vectors[str(i)])

df_vectorized = pd.DataFrame(sequence_of_vectors)
df_vectorized = df_vectorized.reset_index(drop=True)
print(df_vectorized.head())
df_vectorized.to_csv('feb-2014.csv')
# print(len(sequence_of_vectors))
```

```python
window = 500

array = np.array(sequence_of_vectors[:510000])
print(array.shape)

array = np.reshape(array, (len(array)/window, window, 300))
a = [i for i in range(window)]
a = np.reshape(a, (1, window, 1))
sum_of_vectors = np.sum(a*array, axis=1)
print(sum_of_vectors)
print(array.shape)
print(sum_of_vectors.shape)
```

```python
from sklearn.cluster import KMeans


pred = KMeans(n_clusters=42, random_state=13).fit_predict(sum_of_vectors)
pred_df = pd.DataFrame(pred)
print(pred_df.describe())
print(len(pred))
# fig=plt.figure(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k')
print(df.columns)
for i in range(len(sum_of_vectors)):
    if pred[i] == 22:
        print('Index {} - prediction {}'.format(i, pred[i]))
        plt.plot(df[i*window:(i+1)*window])
        plt.show()
        plt.clf()
```
