```python
import pandas as pd
import os 
import numpy as np

year = '2014'
month = '1'

FOLDER_PATH = '/media/christoforos/DATA/signal2vec/vectorized_sequences/{}/{}'.format(year, month)

def calculate_mean_vectors(df):
    window = 500
    array = df.values
    print(array.shape)
    array = np.reshape(array, (len(array)/window, window, 300))
    sum_of_vectors = np.sum(array, axis=1)
    print(array.shape)
    print(sum_of_vectors.shape)
    print('------------------------')
    return sum_of_vectors

en2vec = dict()
for root,dirs,files in os.walk(FOLDER_PATH):
    for f in files:
        if f.endswith(".csv"):
            print(f)
            df = pd.read_csv(FOLDER_PATH + '/' + f)
#             en2vec[f] = calculate_mean_vectors(df)
            en2vec[f] = df.values
            
            
        
```

```python
def calculate_min_max_distances(mains, aggregated):
    samples = mains.shape[0]
    distances = list()
    for i in range(samples):
        dist = np.linalg.norm(mains[i] - aggregated[i])
        distances.append(dist)
    maximum = np.max(distances)
    minimum = np.min(distances)
    mean = np.mean(distances)
    print('Max {}'.format(maximum))
    print('Min {}'.format(minimum))
    print('Mean {}'.format(mean))
    print('-----------------------')


aggregated = None
mains = en2vec['mains.csv']
for k, v in en2vec.items():
    if k is 'mains.csv':
        continue
    if aggregated is None:
        aggregated = v
    else:
        aggregated = np.add(aggregated, v)
    calculate_min_max_distances(mains, aggregated)
#     dist = np.linalg.norm(np.sum(mains, axis=0)-np.sum(aggregated, axis=0))
#     print(dist)

        
```
